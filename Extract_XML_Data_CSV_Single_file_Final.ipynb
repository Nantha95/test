```python
import pandas as pd
import numpy as np
import csv
from csv import writer
from xml.dom import minidom
a = minidom.parse('BIOMD0000000009_url.xml')  # Load single XMLfiles from same directory (either file/directory)
tagname = a.getElementsByTagName('reaction') 
with open("Modelling_CRN_single_file.csv",'w', newline='') as csvfile:
    csv_writer = writer(csvfile)
    csv_writer.writerow(["ID","Reactants","Products","Reversible","Stoichiometry","Stoichiometry_Coefficients","Stoichiometry_Reactant","Stoichiometry_Product"])
    for x in tagname:

#         z1 = z.replace('_url.xml', '')
#         print(z2)
    
        listofReactions=x.getElementsByTagName('listOfReactions')
        tagname = x.getElementsByTagName('reaction')
        species = x.getElementsByTagName('species')
        model_name = x.getElementsByTagName('model')
        
#------------------------------------------------------------------------------------------------------------

        if (x.hasAttribute('id') == True):
            Name_1 = x.attributes['id'].value
#                 Name_2 = m1 + '_' +  Name_1
        else:
            Name_1 = "No name Available"
 #------------------------------------------------------------------------------------------------------------

        b1 = x.getElementsByTagName('listOfReactants')
        reaction = []
        sy = []
        if b1 == []:
            Reactant = ""
        else:
            for x1 in b1:
                c = x1.getElementsByTagName('speciesReference')
                for x2 in c:
                    React_2 = x2.attributes['species'].value
                    Stoic_1 = x2.getAttribute('stoichiometry')
                    if (Stoic_1 == ""):
                        Stoic_1 = 1
                        Stoic_2 = Stoic_1
                    else:
                        Stoic_2 = Stoic_1
                    s = React_2
                    s = reaction.append(s)
                    sy.append(Stoic_2)  
                s2 = " + "
                Reactant = s2.join(reaction)                    
#---------------------------------------------------------------------------------------------------------------

        b2 = x.getElementsByTagName('listOfProducts')
        reaction_2 = []
        sy_2 = []
        if b2 == []:
            Product = ""
        else: 
            for x3 in b2:
                c1 = x3.getElementsByTagName('speciesReference')  
                for x4 in c1:
                    React_4 = x4.attributes['species'].value
                    Stoic_3 = x4.getAttribute('stoichiometry')  
                    if (Stoic_3 == ""):
                        Stoic_3 = 1
                        Stoic_4 = Stoic_3
                    else:
                        Stoic_4 = Stoic_3
                    s1 = React_4
                    s1 = reaction_2.append(s1)
                    sy_2.append(Stoic_4)
                s3 = " + "
                Product = s3.join(reaction_2)
        test = sy+sy_2
        stoic_react = [str(react) for react in sy]
        stoic_prod =  [str(prod) for prod in sy_2]
        stoi = [float(x) for x in test]
#             print(stoi)
        stoic_coeff = stoi
        res = any(ele != 1.0 for ele in stoi)
#             stoichio = res
        if (Reactant == "EmptySet" or Product == "EmptySet"):
            Reactant = Reactant.replace('EmptySet', '')
            Product = Product.replace('EmptySet', '')
        else:
            Reactant = Reactant
            Product = Product
                
        if(Reactant == "" or Product == ""):
            stoichio = True
        else:
            stoichio = res 

    #---------------------------------------------------------------------------------------------------------------

        if(x.getAttribute('reversible'))== "false":
            rev = 1
        
        elif x.getAttribute('reversible') == "true":
            rev = 2
#                         rev_1 = Product
#                         rev_2 = Reactant 
#                         meta_id = Name_2 + "_r"
#                         eq_r = rev_1 + "-->" + rev_2
        else:
            rev = 2
#                         rev_1 = Product 
#                         rev_2 = Reactant 
#                         meta_id = Name_2 + "_r"
#                         eq_r = rev_1 + "-->" + rev_2       
        
        csv_writer.writerow([Name_1,Reactant,Product,rev,stoichio,stoic_coeff,stoic_react,stoic_prod])
```
